# Movie Reccomendation Engine (Hybird Weights Technique)

__In this project we will use our analysis to Tmbd Movies data set to reccommend movies to users based on (popularity and votes) we will use a method called weighted average votes and then combine the average with popularity so thats why it is called hybird it depends on more than one feature in detrmining the reccomendation__

Formula For this is 

$$
  W=\frac{Rv + Cm}{v + m}\
$$

__Where:__
__W:is the weighted Rating__

__R:is average ratings__

__v:The number of votes for the Movie__

__m:Min no of votes to be considered in top (250 Votes)__

__C:Mean vote across the whole report__

```python
r=df['vote_average']
v=df['vote_count']
m=df['vote_count'].quantile(0.80)
c=df['vote_average'].mean()
print(m)
```

```python
df['weighted_average']=((r*v)+(c*m))/(v+m)
```

```python
df.head()
```

```python
movies_sorted=df.sort_values(by='weighted_average',ascending=False)
movies_sorted[['original_title','vote_count','vote_average','weighted_average','popularity']].head(15)
```

__Then you can Reccommend movies based on the weighted average__

__(Movie watched >=recommended movie weighted average)__

__And we can also consider popularity for better results, because we get different movies based on populartiy score so we could use both of them as weights with same priority given to both__

```python
popularity=movies_sorted.sort_values(by='popularity',ascending=False)[0:15]
ax=sns.barplot(x='original_title',y='popularity',data=popularity)
plt.title('Most popular by score of popularity')
plt.xlabel('Movie Title')
ax.tick_params(axis='x', rotation=90)
plt.ylabel('Popularity Score')
plt.show()
```

__Reccomendation Based on Scaled Weighted Average and Popularity Score (50% Priority For both)__

```python
#popularity is 0-200 and weighted avge is between 0 and 1 so we need to normalize the columns
from sklearn.preprocessing import MinMaxScaler

scale=MinMaxScaler()
scaled_df=scale.fit_transform(movies_sorted[['weighted_average','popularity']])
normalized_df=pd.DataFrame(scaled_df,columns=['weighted_average','popularity'])
normalized_df.head()
```

```python
movies_sorted[['norm_weighted_average','norm_popularity']]=normalized_df
```

```python
movies_sorted['final_score']=movies_sorted['norm_weighted_average']*0.50+movies_sorted['norm_popularity']*0.50
final_df_sorted=df.sort_values(by='final_score',ascending=False)
final_df_sorted[['original_title','vote_count','norm_weighted_average','norm_popularity','final_score']].head(20)
```
